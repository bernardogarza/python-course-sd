```python
planet = input('input a planet ')
print(planet)
```

```python
index = planets.index(planet)
```

```python
print(f'Closer to the sun than {planet}:\n{planets[0:index]}')
print(f'Further from the sun than {planet}:\n{planets[index+1:]}')
```
